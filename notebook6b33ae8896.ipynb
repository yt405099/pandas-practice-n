```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/titanic/train.csv")
```

```python
df.head()
```

```python
grouped = df.groupby('Sex')['Age'].agg(['mean', 'std'])
print(grouped)
```

```python
import matplotlib.pyplot as plt
```

```python
grouped['mean'].plot(kind = 'bar', yerr = grouped['std'], capsize = 4)
plt.ylabel('Mean Age')
plt.title('Mean Â± std of Age by Sex')
plt.show()
```
